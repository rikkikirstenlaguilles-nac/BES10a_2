## Program Description

This program checks a student’s possible cognitive state based on:
- caffeine intake  
- study hours  
- stress level  

It uses conditional statements to determine whether the user may be experiencing overstimulation, cognitive fatigue, balanced focus, or low engagement.

This idea is based on my interest in neuroscience and how daily habits affect mental performance as a student.

```python

# Cognitive Fatigue & Focus Checker
# Created by: Kirsten

# This program evaluates a student's possible
# cognitive condition based on caffeine intake,
# study hours, and stress level.

print("Cognitive Fatigue & Focus Checker")

try:
    
    # Asking for the user's name
    name = input("\nEnter your name: ")
    
    # Asking for caffeine intake
    caffeine = int(input("How many cups of coffee did you drink today? "))
    
    # Asking for study hours
    study_hours = int(input("How many hours did you study today? "))
    
    # Asking for stress level
    stress = int(input("Stress level from 1-10: "))
    
    print("\nCognitive Analysis Result:")
    
    # High caffeine and high stress
    if caffeine >= 5 and stress >= 8:
        print("\nPossible overstimulation detected.")
        print("Your brain may be mentally exhausted.")
        print("Consider reducing caffeine intake and resting.")
    
    # Long study hours and high stress
    elif study_hours >= 8 and stress >= 7:
        print("\nHigh risk of cognitive fatigue.")
        print("You may be experiencing mental burnout.")
        print("Take breaks and get proper sleep.")
    
    # Balanced productivity
    elif study_hours >= 4 and stress <= 5:
        print("\nHealthy productivity level detected.")
        print("Your focus condition appears balanced.")
    
    # Low study hours
    elif study_hours < 2:
        print("\nLow cognitive engagement detected.")
        print("Try improving your study consistency.")
    
    # Default condition
    else:
        print("\nModerate mental strain detected.")
        print("Try managing stress and resting properly.")
    
    # Thank you message
    print("\nThank you for using the program,", name + "!")

# Handles invalid inputs
except:
    print("\nInvalid input.")
    print("Please enter numbers only.")

```
